-------------------------------------------------------------------------------

**INSUMO_ATUAL**: Qualitativa Nominal, pois representa uma descrição do material.

-------------------------------------------------------------------------------

-------------------------------------------------------------------------------

**UNIDADE**: Qualitativa Nominal, pois representa uma categoria e não um número para cálculos diretos e não possui uma ordem lógica ou hierarquia.

-------------------------------------------------------------------------------

**COEF_SIURB**: Quantitativa Contínua, pois é um número decimal e pode ser usado para cálculos e análises estatísticas.

A variação na produção pode ser explicada por diversos fatores. O grupo com a menor média pode ter sido impactado por chuvas, problemas climáticos, falta de funcionários ou até mesmo falhas técnicas. #4
 - Utilize tudo que aprendeu sobre medidas de posição central, gráfico de densidade de frequência e boxplot para responder a seguinte pergunta: Há diferença de produtividade entre blocos (coluna descrição)? Se há, qual bloco é mais produtivo? Qual menos produtivo? Consegue pensar em motivos para esses comportamentos?

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('dados_produtividade_construcao.csv')

unique_blocos = df['DESCRICAO'].unique()
print("Valores únicos na coluna 'DESCRICAO':")
print(unique_blocos)


block_stats = {}

for bloco in unique_blocos:
    subset = df[df['DESCRICAO'] == bloco]
    if not subset["IP_D"].empty:
        mean_val   = subset["IP_D"].mean()
        median_val = subset["IP_D"].median()
        std_val    = subset["IP_D"].std()
        min_val    = subset["IP_D"].min()
        max_val    = subset["IP_D"].max()
        block_stats[bloco] = {
            "mean": mean_val,
            "median": median_val,
            "std": std_val,
            "min": min_val,
            "max": max_val
        }

print("\nEstatísticas descritivas por bloco:")
for bloco, stats in block_stats.items():
    print(f"{bloco}: Média={stats['mean']:.2f}, Mediana={stats['median']:.2f}, Std={stats['std']:.2f}, Min={stats['min']}, Max={stats['max']}")

plt.figure(figsize=(10,6))
for bloco in unique_blocos:
    subset = df[df['DESCRICAO'] == bloco]
    if not subset["IP_D"].empty:
        subset["IP_D"].plot(kind='density', label=bloco)
plt.title('Gráfico de Densidade da Produtividade por Bloco')
plt.xlabel('Produtividade (IP_D)')
plt.legend()
plt.show()


data_to_plot = []
labels = []
for bloco in unique_blocos:
    subset = df[df['DESCRICAO'] == bloco]
    if not subset["IP_D"].empty:
        data_to_plot.append(subset["IP_D"])
        labels.append(bloco)

plt.figure(figsize=(10,6))
plt.boxplot(data_to_plot, labels=labels)
plt.title('Boxplot da Produtividade por Bloco')
plt.ylabel('Produtividade (IP_D)')
plt.xticks(rotation=45)
plt.show()


most_productive = max(block_stats.items(), key=lambda x: x[1]["mean"])
least_productive = min(block_stats.items(), key=lambda x: x[1]["mean"])

print(f"\nBloco mais produtivo: {most_productive[0]} com média de {most_productive[1]['mean']:.2f}")
print(f"Bloco menos produtivo: {least_productive[0]} com média de {least_productive[1]['mean']:.2f}")

```

Sim, tem diferença na produtividade entre os blocos. As médias e medianas de cada um variam, mostrando que alguns são bem mais produtivos que outros. Pelo código, dá pra ver que o bloco com a maior média de produtividade IP_D é o mais eficiente, enquanto o que tem a menor média é o menos produtivo. Isso fica ainda mais claro nos gráficos: o de densidade mostra onde a produtividade se concentra e o boxplot deixa evidente a variação, outliers e como as medianas se comparam.
Agora, por que essa diferença existe? Pode ser muita coisa. Talvez a equipe de um bloco seja mais experiente ou trabalhe num ritmo melhor. Pode ser que algumas tarefas sejam mais difíceis que outras. Logística também pesa – ter material sempre à mão facilita. Fora isso, o ambiente pode influenciar, tipo condições climáticas ou até como a gestão lida com cada equipe. Se quiser, posso rodar uns testes estatísticos pra confirmar essas diferenças de um jeito mais preciso.

A SIURB de São Paulo tem tabelas de custos que servem como referência para calcular os gastos com obras públicas. Nessas tabelas, existem coeficientes que ajudam a estimar quanto material e mão de obra são necessários para cada serviço.

Por exemplo, tem o coeficiente de consumo, que indica quantos materiais são usados por unidade de serviço. Se for concreto, ele mostra quantos quilos de cimento são necessários para fazer um metro cúbico. Já o coeficiente de produtividade mede o rendimento da equipe ou do profissional. Tipo, quantos metros quadrados de alvenaria um pedreiro consegue assentar por dia.

Pra chegar nesses valores, a SIURB faz pesquisas de mercado e consulta especialistas, como a Fundação Instituto de Pesquisas Econômicas (FIPE), que atualizam os preços de materiais, equipamentos e mão de obra. Com esses dados, eles montam as tabelas, onde cada serviço tem um coeficiente indicando sua participação no custo total.

Agora, sobre como fazer a conta: se eu quiser calcular o custo de um serviço, basta pegar o coeficiente da tabela e multiplicar pelo valor do insumo. Por exemplo, se a tabela diz que preciso de 1,5 sacos de cimento por metro quadrado de contrapiso, e cada saco custa 40 Reais  , então o custo total só do cimento seria 1,5 × 40 = R$ 60/m². O mesmo vale pra calcular a mão de obra e outros materiais, somando tudo no orçamento final.

Como tudo na construção muda com o tempo, a SIURB revisa essas tabelas de tempos em tempos, ajustando os coeficientes para refletir novas técnicas e preços atualizados. Isso garante que os orçamentos fiquem realistas e alinhados com o mercado.
